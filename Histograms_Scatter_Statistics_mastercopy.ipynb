```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the data from the Google Sheets link
url = "https://docs.google.com/spreadsheets/d/15JMfQmqYunDQFQnGIcJYHd6P923kaK1k7EndvM2wWWk/export"
df = pd.read_excel(url)

# Drop rows with missing values
df.dropna(subset=['Year','Mass (g)', ], inplace=True)

# Convert 'year' column to numeric type (removing non-numeric values)
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Mass (g)'], weights=df['Year'], bins=100, edgecolor='black')
plt.xlabel('Mass (g)')
plt.ylabel('Year')
plt.title('Histogram of Mass (g) by Year')
plt.grid(True)
plt.show()
```

##Muons!!##
This is from the Departmrnt of Energy website, https://www.energy.gov/science/doe-explainsmuonsn The muon is one of the fundamental subatomic particles, the most basic building blocks of the universe as described in the Standard Model of particle physics. Muons are similar to electrons but weigh more than 207 times as much. The muon is part of the lepton group. Leptons are a type of fundamental particle. This means they are not made of even smaller pieces of matter. Like other leptons, the muon is affected by only three of the four fundamental forces in the universe.

The muons that hit the Earth result from particles in the Earths atmosphere colliding with cosmic rayshigh-energy protons and atomic nuclei that move through space at just below the speed of light. Muons exist for only 2.2 microseconds before they decay into an electron and two kinds of neutrinos. However, because they move at nearly the speed of light, muons travel far before decaying. Muons created in the atmosphere constantly hit every inch of the Earths surface and pass through almost any substance. They dont stop until they penetrate far below the surface of the Earthpotentially more than a mile.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://docs.google.com/spreadsheets/d/1te6fVgTtUiSpVTpMU8yPxeUJLXkE5fcZ0GprwvfmmlQ/export"
df = pd.read_excel(url)

# Drop rows with missing values
df.dropna(subset=['Time (ns)','Distance (m)', ], inplace=True)

# Convert 'year' column to numeric type (removing non-numeric values)
# df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Time (ns)'], weights=df['Distance (m)'], bins=100, edgecolor='black')
plt.xlabel('Distance (m)')
plt.ylabel('Time (ns)')
plt.title('Histogram of Time (ns) by Distance (m)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
Polynomial = np.polynomial.Polynomial

xdata = df['Time (ns)'].values
ydata = df['Distance (m)'].values

plt.figure(figsize=(10, 6))
plt.scatter(xdata, ydata)
plt.ylabel('Distance (m)')
plt.xlabel('Time (ns)')
plt.title('Graph of Distance (m) as a functon of Time of Flight (ns)')
plt.grid(True)
m, b = np.polyfit(xdata, ydata, 1)
print(m)
# adding the regression line to the scatter plot
plt.plot(xdata, m*xdata + b)
plt.show()
slope, intercept = np.polyfit(np.log(xdata), np.log(ydata), 1)
# print(slope)

print('x-mean = {:.2f}, x-var = {:.2f}'.format(xdata.mean(), xdata.var()))
print('y-mean = {:.2f}, y-var = {:.2f}'.format(ydata.mean(), ydata.var()))
# The correlation coefficient between (x, y) is the off-diagonal element
# of the correlation coefficient matrix
corrcoef = np.corrcoef(xdata, ydata)[0][1]
print('(x, y) correlation coefficient = {:.3f}'.format(corrcoef))
# Fit a straight line to the data
pfit = Polynomial.fit(xdata, ydata, 1, window=(xdata.min(), xdata.max()))
print('Linear regression: y = {:.2f}x + {:.2f}'.format(*pfit.coef[::-1]))
# fitting a linear regression line
```
